# Package Import Smoke

A tiny notebook used by pytest to verify that notebook code cells can execute against the installed package surface.

```python
import common
import qite
import qpe
import vqe

assert common.__name__ == "common"
assert qite.__name__ == "qite"
assert qpe.__name__ == "qpe"
assert vqe.__name__ == "vqe"
```

```python
import pennylane as qml

from common.persist import canonical_hamiltonian
from vqe.auto_ansatz import resolve_auto_ansatz

hamiltonian = qml.Hamiltonian([1.0], [qml.PauliZ(0)])
fingerprint = canonical_hamiltonian(hamiltonian)
selected, kwargs, metadata = resolve_auto_ansatz("auto", hamiltonian, 1)

assert fingerprint["pauli_terms"] == [
    {"ops": [[0, "Z"]], "coeff_real": 1.0, "coeff_imag": 0.0}
]
assert selected == "StronglyEntanglingLayers"
assert kwargs == {"layers": 2}
assert metadata["requested"] == "auto"
```
