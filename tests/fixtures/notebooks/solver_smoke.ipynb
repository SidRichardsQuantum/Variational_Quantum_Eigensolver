# Solver Smoke

A tiny executed notebook fixture that exercises expert-mode solver paths without invoking chemistry backends.

```python
import pennylane as qml

hamiltonian = qml.Hamiltonian([1.0], [qml.PauliZ(0)])
reference_state = [1]
```

```python
from vqe import run_vqe

vqe_result = run_vqe(
    molecule="notebook_expert_z",
    hamiltonian=hamiltonian,
    num_qubits=1,
    reference_state=reference_state,
    ansatz_name="RY-CZ",
    steps=1,
    seed=0,
    plot=False,
    force=True,
)

assert "energy" in vqe_result
assert vqe_result["num_qubits"] == 1
```

```python
from qite import run_qite

qite_result = run_qite(
    molecule="notebook_expert_z",
    hamiltonian=hamiltonian,
    num_qubits=1,
    reference_state=reference_state,
    ansatz_name="RY-CZ",
    steps=1,
    dtau=0.01,
    seed=0,
    force=True,
)

assert "energy" in qite_result
assert qite_result["num_qubits"] == 1
```

```python
from qpe import run_qpe

qpe_result = run_qpe(
    molecule="notebook_expert_z",
    hamiltonian=hamiltonian,
    hf_state=reference_state,
    system_qubits=1,
    n_ancilla=1,
    t=0.1,
    trotter_steps=1,
    shots=None,
    seed=0,
    force=True,
)

assert "energy" in qpe_result
assert qpe_result["num_qubits"] == 1
```
