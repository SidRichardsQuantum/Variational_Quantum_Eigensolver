# Atomic Ionization Energy Benchmark

This notebook expands benchmark coverage beyond the H2/LiH molecular panel by using the shared registry to compare neutral atoms with their cation partners.

It reports:

- exact neutral and cation ground-state energies from the standard qubit-Hamiltonian pipeline
- ionization energy in Hartree and eV
- electron, qubit, and Hamiltonian-term counts for each side of the pair

The notebook is a decision benchmark rather than a usage tutorial. It is intended to show where the registry already supports same-pipeline atomic reference studies and where qubit or term-count growth starts to matter.

```python
from __future__ import annotations

import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display

from common import ionization_energy_panel, summarize_registry_coverage

UNIT = "angstrom"
MAPPING = "jordan_wigner"
SYSTEMS = ["He", "He+", "Be", "Be+", "B", "B+", "C", "C+", "N", "N+", "O", "O+", "F", "F+"]
PAIRS = [("He", "He+"), ("Be", "Be+"), ("B", "B+"), ("C", "C+"), ("N", "N+"), ("O", "O+"), ("F", "F+")]
```

```python
coverage_rows = summarize_registry_coverage(
    systems=SYSTEMS,
    mapping=MAPPING,
    unit=UNIT,
    progress=True,
)

ionization_rows = ionization_energy_panel(coverage_rows, pairs=PAIRS)
summary_df = pd.DataFrame(ionization_rows).sort_values(
    ["neutral_num_qubits", "ionization_energy_ev", "system"]
)

display(
    summary_df[
        [
            "system",
            "neutral",
            "cation",
            "ionization_energy_ha",
            "ionization_energy_ev",
            "neutral_exact_ground_energy",
            "cation_exact_ground_energy",
            "neutral_num_electrons",
            "cation_num_electrons",
            "neutral_num_qubits",
            "cation_num_qubits",
            "neutral_hamiltonian_terms",
            "cation_hamiltonian_terms",
        ]
    ].round(8)
)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 4.8))

axes[0].bar(summary_df["system"], summary_df["ionization_energy_ev"], color="#355070")
axes[0].set_title("Ionization Energy")
axes[0].set_ylabel("eV")
axes[0].grid(True, axis="y", alpha=0.3)

x = range(len(summary_df))
width = 0.38
axes[1].bar([i - width / 2 for i in x], summary_df["neutral_hamiltonian_terms"], width=width, label="neutral", color="#6d597a")
axes[1].bar([i + width / 2 for i in x], summary_df["cation_hamiltonian_terms"], width=width, label="cation", color="#b56576")
axes[1].set_xticks(list(x), summary_df["system"])
axes[1].set_title("Hamiltonian Term Counts")
axes[1].set_ylabel("terms")
axes[1].legend()
axes[1].grid(True, axis="y", alpha=0.3)

fig.suptitle("Registry atomic ionization benchmark")
fig.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
