# Registry Coverage

This notebook summarizes the built-in molecule registry used by the shared chemistry pipeline.

It reports:

- registry molecule name
- charge and spin multiplicity defaults
- resolved electron and qubit counts
- Hamiltonian term count
- exact ground-state reference energy

The notebook is meant to be a decision notebook, not a tutorial. It uses the same `build_hamiltonian(...)` path that powers the packaged solver workflows, so the table reflects actual supported small-system coverage rather than a separate hand-maintained list.

```python
from __future__ import annotations

import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display

from common import summarize_registry_coverage

UNIT = "angstrom"
MAPPING = "jordan_wigner"
REGISTRY_SYSTEMS = None
```

```python
rows = summarize_registry_coverage(
    systems=REGISTRY_SYSTEMS,
    mapping=MAPPING,
    unit=UNIT,
    progress=True,
)

summary_df = pd.DataFrame(rows).sort_values(
    ["num_qubits", "num_electrons", "charge", "molecule"]
)

display(
    summary_df[
        [
            "molecule",
            "symbols",
            "charge",
            "multiplicity",
            "basis",
            "num_electrons",
            "num_qubits",
            "hamiltonian_terms",
            "exact_ground_energy",
        ]
    ].round(8)
)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 4.8))

axes[0].bar(summary_df["molecule"], summary_df["num_qubits"], color="#355070")
axes[0].set_title("Resolved Qubits by Registry Molecule")
axes[0].set_ylabel("Qubits")
axes[0].tick_params(axis="x", rotation=90)
axes[0].grid(True, axis="y", alpha=0.3)

axes[1].bar(summary_df["molecule"], summary_df["hamiltonian_terms"], color="#b56576")
axes[1].set_title("Hamiltonian Terms by Registry Molecule")
axes[1].set_ylabel("Term count")
axes[1].tick_params(axis="x", rotation=90)
axes[1].grid(True, axis="y", alpha=0.3)

fig.suptitle("Shared chemistry-registry coverage")
fig.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
focus_view = summary_df[["molecule", "num_electrons", "num_qubits", "hamiltonian_terms"]].copy()
focus_view = focus_view.sort_values(["num_qubits", "hamiltonian_terms", "molecule"])
focus_view.round(8)
```
